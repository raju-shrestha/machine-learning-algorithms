Predicting the Breast Caner Survial patient rate from the patient having cancer of different age group, menopause, tumor size, breast part using Logistic Regression algorithm. 

```python
# Import pandas 
import pandas as pd
```

Load the dataframe in pandas from csv file

```python
df = pd.read_csv('data/breast_cancer.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
label = df["Class"]
```

```python
label.head()
```

```python
# Droping Unnecessary columns
features = df.drop("Class", axis=1)
features = features.drop("Inv_Nodes", axis=1)
```

```python
features.head()
```

```python
# Taking age as mid value
age = {"20-29":25, "30-39":35, "40-49":45, "50-59":55, "60-69":65, "70-79":75}
features["Age"] = features["Age"].map(age)
features.head(200)
```

```python
# Taking '?' as "no'
features["Node_Caps"] = features["Node_Caps"].replace("?", "no")
```

```python
# Taking '?' as "no'
features["Breast_Quad"] = features["Breast_Quad"].replace("?", "left_low")
```

```python
# Creating dummy variable for Menopause, Node_Caps, Deg_Malig, Breast, Breast_Quad and Irradiat
features = pd.get_dummies(features, columns=["Menopause", "Node_Caps", "Deg_Malig", "Breast", "Breast_Quad", "Irradiat"])
```

```python
features.head(20)
```

```python
# Counting Tumor size
features["Tumor_Size"].value_counts()
```

```python
# Replacing the  String date to number value
features["Tumor_Size"] = features["Tumor_Size"].replace("14-Oct", "10-14")
features["Tumor_Size"] = features["Tumor_Size"].replace("9-May", "5-9")
```

```python
# Taking mid value for Tumor size
tumor_size = {"0-4":2, "5-9":7, "10-14":12, "15-19":17, "20-24":22, "25-29":27, "30-34":32, "35-39":37, "40-44":42, "45-49":47, "50-54":52}
features["Tumor_Size"] = features["Tumor_Size"].map(tumor_size)
features.head(20)
```

```python
features.head()
```

```python
features.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(features, label, test_size=0.3, random_state=60)
```

```python
from sklearn.linear_model import LogisticRegression
regressor = LogisticRegression()
```

```python
regressor.fit(x_train, y_train)
```

```python
prediction = regressor.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(prediction, y_test)*100
```

Accuracy 76.74 %
